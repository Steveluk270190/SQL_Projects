# Task 2: Sales Analysis

```python

#Calculate the total revenue generated by each Order in the OrderDetail table
#revenue=SUM(SUM(OD.quantity* OD.unitPrice)).
#If the OD.quantity* OD.unitPrice is associated with only 1 SUM(), then it would result in the value of each order within a single day only instead of the value of different orders across different days in addition to the value of each order within a single day.

t_revenue='''With cte_1 AS (SELECT OD.orderId, SUM(OD.quantity*OD.unitPrice -OD.discount) AS ordervalue
          FROM OrderDetail as OD
          GROUP BY OD.orderId)
      SELECT cte_1.orderId, SUM(cte_1.ordervalue) AS revenue
      FROM cte_1
      GROUP BY 1;'''
t_revenue= pd.read_sql_query(t_revenue, conn)
print(t_revenue)
```

```python

# Identify the top 5 customers who made the highest number of orders

top_5_customers = ''' SELECT SalesOrder.customerId, COUNT(orderId) AS no_of_orders
                     FROM SalesOrder JOIN OrderDetail
                     ON OrderDetail.orderId=SalesOrder.entityId
                     GROUP BY SalesOrder.customerId
                     ORDER BY no_of_orders DESC
                     LIMIT 5;'''
top_5_customers = pd.read_sql_query(top_5_customers, conn)
print(top_5_customers)
```

```python

#Determine the top 5 best-selling products along with their quantities sold

best_selling_products ='''
  SELECT PT.entityId, SUM(OD.quantity) as quantities_sold
  FROM Product AS PT JOIN OrderDetail AS OD
  ON PT.entityId = OD.productId
  GROUP BY PT.entityId
  ORDER BY quantities_sold DESC
  LIMIT 5;'''
best_selling_products= pd.read_sql_query(best_selling_products, conn)
print(best_selling_products)
```

# Task 3: Customer Insights

```python

#Calculate the average order value (AOV) across all orders per each customer.

AOV='''With cte_1 AS (SELECT OD.orderId, SO.customerId, SUM(OD.quantity*OD.unitPrice-OD.discount) AS Revenue
                      FROM OrderDetail as OD
                      JOIN
                      SalesOrder as SO ON SO.entityId=OD.orderId
                      GROUP BY 1, 2)
      SELECT cte_1.customerId, SUM(cte_1.Revenue)/(COUNT(*)) AS AOV
      FROM cte_1
      GROUP BY 1'''
AOV= pd.read_sql_query(AOV, conn)
print(AOV)

```

```python

#Find the total number of order, and the average price of order per each customer:
#count distinct becuz we want each customer to receive only 1 count

avg_no_order='''SELECT SO.customerId, COUNT(DISTINCT OD.orderId) AS Count_order, SUM(OD.unitprice)/(COUNT(DISTINCT OD.orderId)) as Avg_order_price
                                FROM OrderDetail as OD
                                JOIN
                                SalesOrder as SO ON SO.entityId=OD.orderId
                                GROUP BY 1'''
avg_no_order= pd.read_sql_query(avg_no_order, conn)
print(avg_no_order)
```

```python

#Find the total order quantity and the average order quantity per each customer

average_quantity='''SELECT SO.customerId, SUM(OD.quantity) AS Sum_order_quantity,AVG(OD.quantity) AS Average_order_quantity
               FROM SalesOrder as SO
               JOIN OrderDetail as OD ON SO.entityId=OD.orderId
              GROUP BY 1;'''
average_quantity = pd.read_sql_query(average_quantity, conn)
print(average_quantity)



```

```python

# Find the customer who has the highest total expenditure on purchases.
#Total Expenditure = SUM(SUM(OD.quantity* OD.unitprice))

Total_Expenditure = '''With cte_1 AS (SELECT SO.customerId, OD.orderId, SUM(OD.quantity* OD.unitprice-OD.discount) AS revenue
                                      FROM OrderDetail as OD
                                      JOIN
                                      SalesOrder as SO ON SO.entityId=OD.orderId
                                      GROUP BY 1,2
                                      ORDER BY 1)
                     SELECT cte_1.customerId, SUM(cte_1.revenue) AS total_expenditure
                     FROM cte_1
                     GROUP BY 1
                     ORDER BY total_expenditure DESC
                     LIMIT 1;'''
Total_Expenditure = pd.read_sql_query(Total_Expenditure, conn)
print(Total_Expenditure)

```

```python

#Identify the customers with a large total expenditure (>$10,000), a medium-sized total expenditure ($1,000 - $10,000), and a small total expenditure ($1 - $1,000) on their orders.

Expenditure = '''With cte_1 AS (SELECT OD.orderId, SO.customerId, SUM(OD.quantity* OD.unitprice-OD.discount) AS revenue
                                      FROM OrderDetail as OD
                                      JOIN
                                      SalesOrder as SO ON SO.entityId=OD.orderId
                                      GROUP BY 1, 2
                                      )
                   SELECT cte_1.customerId, CASE WHEN SUM(cte_1.revenue) > 10000 THEN 'Large_expenditure'
                                                WHEN SUM(cte_1.revenue) >= 1000 AND SUM(cte_1.revenue) <= 10000 THEN 'Medium-sized_expenditure'
                                                ELSE 'Small_expenditure' END AS expenditure_category
                  FROM cte_1
                  GROUP BY 1;'''
Expenditure = pd.read_sql_query(Expenditure, conn)
print(Expenditure)








```

```python

# Find the count of each country's order for each revenue category

revenue_country = '''
WITH cte_1 AS (
    SELECT OD.orderId, SO.shipCountry, SUM(OD.quantity * OD.unitprice-OD.discount) AS revenue
    FROM OrderDetail AS OD
    JOIN SalesOrder AS SO ON SO.entityId = OD.orderId
    GROUP BY 1, 2
),
cte_2 AS (
    SELECT cte_1.shipCountry,
    CASE
        WHEN SUM(cte_1.revenue) > 10000 THEN 'Large_expenditure'
        WHEN SUM(cte_1.revenue) >= 1000 AND SUM(cte_1.revenue) <= 10000 THEN 'Medium-sized_expenditure'
        ELSE 'Small_expenditure'
    END AS revenue_category
    FROM cte_1
    GROUP BY cte_1.shipCountry)
SELECT cte_2.shipCountry,
       cte_2.revenue_category,
       COUNT(cte_1.orderId) OVER (PARTITION BY cte_2.shipCountry, cte_2.revenue_category) as Country_order_count
       FROM cte_1
       JOIN cte_2 ON cte_1.shipCountry=cte_2.shipCountry '''

revenue_country = pd.read_sql_query(revenue_country, conn)
print(revenue_country)
```

```python

#Calculate the total revenue generated by each category.

total_revenue='''
With cte_1 AS (SELECT PT.CategoryId, SUM(OD.quantity* OD.unitprice-OD.discount) AS revenue
          FROM OrderDetail as OD
          JOIN Product as PT ON Pt.entityId=OD.productId
          GROUP BY 1)
      SELECT cte_1.categoryId, SUM(cte_1.revenue) AS total_revenue
      FROM cte_1
      GROUP BY 1;'''
total_revenue = pd.read_sql_query(total_revenue, conn)
print(total_revenue)
```

```python

#Find the supplier who has the highest number of products' in each category. (Using Window function)

#Partitioning by the categoryId only in cte_2:
###multiple suppliers provide products in the same category, Dense_Rank will identify ALL the suppliers with the HIGHEST product_count across ALL suppliers within that category.
###If partitioning by BOTH categoryId and supplierId, then the Dense_Rank will find the supplier with the highest Product_Count for "each supplier" and "each category".
###Given that the supplierId is unique for every supplier and the "highest ProductCount" of "each supplier" is "that supplier alone", the result will show ,instead of the suppliers with the highest ProductCount in each category,the "highest ProductCount" of "each and every supplier" in "the entire dataset"
###which means ,in other words, the "ProductCount" of "each and every supplier" in "the entire dataset"

CategoryProductCounts='''
WITH cte_1 AS (SELECT PT.categoryId, PT.supplierId,
                COUNT(PT.entityId) AS ProductCount
                FROM  Product as PT
                GROUP BY 1,2),
    cte_2 AS (
            SELECT
                cte_1.categoryId,
                cte_1.supplierId,
                (cte_1.ProductCount) AS Total_ProductCount,
                DENSE_RANK() OVER (PARTITION BY cte_1.categoryId ORDER BY (cte_1.ProductCount) DESC) AS Product_rank
            FROM
                cte_1)
SELECT
    cte_2.categoryId,
    cte_2.supplierId,
    cte_2.Total_ProductCount,
    cte_2.Product_rank
FROM
    cte_2
WHERE
    cte_2.Product_rank = 1
'''

CategoryProductCounts = pd.read_sql_query(CategoryProductCounts, conn)
print(CategoryProductCounts)
```

# Task 6: Aggregation and Grouping

```python

#Calculate the total revenue, average order value, and total number of orders for each country in the Customers table.
###### orderId should be included in cte_1 since the orderId is required in the main query

Aggregation = '''With cte_1 AS (SELECT OD.orderId, C.country, SUM(OD.quantity* OD.unitprice-OD.discount) AS order_value
                              FROM OrderDetail as OD
                              JOIN SalesOrder as SO ON OD.orderId=SO.entityId
                              JOIN Customer as C ON SO.customerId=C.entityId
                              GROUP BY 1,2)
SELECT cte_1.country, SUM(cte_1.order_value) AS total_revenue, COUNT(DISTINCT cte_1.orderId) as Total_order, SUM(cte_1.order_value)/COUNT(*) as Average_Order_Value
FROM cte_1
GROUP BY 1;'''

Aggregation = pd.read_sql_query(Aggregation, conn)
print(Aggregation)



```

```python

#Determine the total quantity of each product sold, grouped by customer.
#total quantity=SUM(SUM(OD.quantity))

Aggregation_1a = '''WITH cte_1 AS (SELECT OD.productId,SO.customerId,SUM(OD.quantity) as SUM_quantity
                                  FROM OrderDetail AS OD
                                  JOIN SalesOrder as SO
                                  ON OD.orderId=SO.entityId
                                  GROUP BY 1,2)
SELECT  cte_1.productId,
        cte_1.customerId,
        SUM(cte_1.SUM_quantity) OVER (PARTITION BY cte_1.productId, cte_1.customerId) AS total_product_quantity_each_customer
        FROM cte_1;'''
Aggregation_1a = pd.read_sql_query(Aggregation_1a, conn)
print(Aggregation_1a)
```

```python

#Determine the total quantity of each product sold, grouped by customer.(Non-window function equivalent)

Aggregation_4 = '''With cte_1 AS (SELECT OD.productId, SO.customerId, SUM(OD.quantity) AS total_quantity_of_each_product_sold
                                 FROM SalesOrder AS SO
                                 JOIN
                                 OrderDetail as OD ON OD.orderId = SO.entityId
                                 GROUP BY 1,2)
                  SELECT cte_1.productId, cte_1.customerId, SUM(cte_1.total_quantity_of_each_product_sold) AS total_product_quantity_each_customer
                  FROM cte_1
                  GROUP BY 1,2;'''

Aggregation_4 = pd.read_sql_query(Aggregation_4 , conn)
print(Aggregation_4 )
```

```python

#Calculate the total revenue generated by each employee for the year 2007.

Aggregation_5 = '''With cte_1 AS (SELECT SO.employeeId, SO.orderDate, SUM(OD.quantity* OD.unitprice-OD.discount) AS order_value
                              FROM OrderDetail as OD
                              JOIN SalesOrder as SO ON OD.orderId=SO.entityId
                              GROUP BY 1,2)
SELECT cte_1.employeeId, SUM(cte_1.order_value) AS total_revenue
FROM cte_1
WHERE cte_1.orderDate >= '2007-01-01' AND cte_1.orderDate <= '2007-12-31'
GROUP BY 1;'''

Aggregation_5 = pd.read_sql_query(Aggregation_5 , conn)
print(Aggregation_5 )
```

# Task 7: Advanced Analysis

```python

#Identify the order_date and the customers who have who have placed more than 1 order on the same day

customers_win='''WITH cte_1 as (SELECT  SO.orderDate,
                                    SO.customerId,
                                    COUNT(OD.orderId) OVER (PARTITION BY SO.orderDate, SO.customerId) as total_count
                                    FROM SalesOrder as SO
                                    JOIN OrderDetail as OD ON OD.orderId=SO.entityId)
          SELECT cte_1.orderDate, cte_1.customerId, cte_1.total_count
          FROM cte_1
          GROUP BY 1,2
          HAVING cte_1.total_count>1;'''

customers_win = pd.read_sql_query(customers_win , conn)
print(customers_win )

```

```python

#Identify the order_date and the customers who have who have placed more than one order on the same day-----The Non-WINDOW function alternative

customers_n='''WITH cte_1 as (SELECT  SO.orderDate,
                                    SO.customerId,
                                    COUNT(OD.orderId) as total_count
                                    FROM SalesOrder as SO
                                    JOIN OrderDetail as OD ON OD.orderId=SO.entityId
                                    GROUP BY 1,2)
            SELECT cte_1.orderDate, cte_1.customerId, cte_1.total_count
                    FROM cte_1
                    GROUP BY 1,2
                    HAVING cte_1.total_count > 1;'''

customers_n = pd.read_sql_query(customers_n , conn)
print(customers_n)

```

```python

#Calculate the average time taken between order date and shipped date

avg_days_taken = '''
    WITH cte_1 as (SELECT OD.orderId, AVG(JULIANDAY(SO.shippedDate) - JULIANDAY(SO.orderDate)) as avg_days_taken
    FROM OrderDetail as OD
    JOIN SalesOrder as SO ON OD.orderId = SO.entityId
    GROUP BY 1)
SELECT cte_1.orderId, cte_1.avg_days_taken
FROM cte_1
GROUP BY 1;
'''

avg_days_taken = pd.read_sql_query(avg_days_taken , conn)
print(avg_days_taken)
```

```python

# Find the most popular pair of products that are often ordered together.
###WHERE OD1.productId < OD2.productId: Ensure we don't count pairs twice (can be both "<"/">")

popular_pair = '''
WITH cte_1 AS (
    SELECT OD1.orderId, OD2.orderId, OD1.productId AS product1, OD2.productId AS product2
    FROM OrderDetail AS OD1
    JOIN OrderDetail AS OD2 ON OD1.orderId = OD2.orderId
    WHERE OD1.productId < OD2.productId
)
SELECT cte_1.product1, cte_1.product2, COUNT(cte_1.orderId) AS order_count
FROM cte_1
GROUP BY 1, 2
ORDER BY order_count DESC
LIMIT 1;
'''

popular_pair = pd.read_sql_query(popular_pair , conn)
print(popular_pair)
```
